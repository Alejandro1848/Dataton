Como se ve en la gráfica anterior, existen valores que pertenecen a la misma categoría, pero que por diferir en caracteres los clasifica como si fueran diferentes. Por lo que se tendrá que hacer limpieza de datos respecto a esto.

Para ver en que se diferencian los caracteres(string), podemos consultar los valores que toman estas categorias y luego proceder a hacer la limpieza de datos.

```python
df.Categoria.unique()
```

De la línea de código anterior podemos observar que por ejemplo de repiten tres veces los caracteres 'Programa/Acción Social'
con la diferencia de que uno tiene un espacio despues de la palabra social: 'Programa/Acción Social ' y el otro tiene la S en minúscula:  'Programa/Acción social'. Algo similar pasa con el string: 'Medidas de Contención ' que tiene un espacio al final. Para hacer la limpieza de datos basta con sustituir cada uno de estos valores incorrectos por el correcto, como se hace en las siguiente líneas de código:

```python
df['Categoria'] = df['Categoria'].str.replace('Medidas de Contención ' ,'Medidas de Contención')
```

```python
df['Categoria'] = df['Categoria'].str.replace('Programa/Acción Social ' ,'Programa/Acción Social')
```

```python
df['Categoria'] = df['Categoria'].str.replace('Programa/Acción social' ,'Programa/Acción Social')
```

```python
sns.set(rc={'figure.figsize':(14.7,11.27)})
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Categoria", data=df)
```

Ahora sí estos eran los gráficos que esperábamos ver. Observamos que las medidas de contención son equiparables en cuanto a cantidad respecto a los programas y acciones sociales que se han llevado a cabo de manera paralela durtante la pandemia.

Haremos ahora un nuevo data set que contenga solo aquellos datos relacionados con medidas de contención en la columna de categoría:

```python
in_contencion= df["Categoria"]=="Medidas de Contención"
```

```python
df_contencion=df.loc[in_contencion]
```

```python
df_contencion.shape
```

```python
df_contencion.Medida.values
```

```python
fecha_categoria = df.groupby(["Fecha_anuncio","Categoria"]).size()
fecha_categoria
```

```python
fecha_categoria.values
```

```python

sns.set(rc={'figure.figsize':(20.7,10.27)})
fig, ax = plt.subplots(figsize=(15, 15))
fecha_categoria.plot(kind='area',xlabel='Fecha_anuncio')
ax.set(xlabel="Fecha",
       ylabel="Cantidad",
       title="Historial de casos positivos diarios desde 2020-03-01 al 2021-07-02 (AA/MM/DD)",
       )

plt.show()
```

# Cantidad solo de medidas de contención, ordenadas por fecha.

```python
fecha_cont = df_contencion.groupby(["Fecha_anuncio","Categoria"]).size()
```

```python
sns.set(rc={'figure.figsize':(20.7,10.27)})
sns.set(font_scale = 1.5)
fig, ax = plt.subplots(figsize=(10, 10))
fecha_cont.plot(kind='area',xlabel='Fecha_anuncio')

ax.set(xlabel="Fecha",
       ylabel="Medidas de contención",
       title="Cantidad de medidas de contención anunciadas por fecha",
       )
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.xlabel("Fecha", size=16)
plt.ylabel("Medidas de contención", size=16)
plt.title("Cantidad de medidas de contención anunciadas por fecha", size=24)
plt.savefig('Graph1_contencion.png', dpi=300)
plt.show()

```
